```python
#Import required Image library
from PIL import Image, ImageFilter, ImageEnhance
import matplotlib.pyplot as plt
import matplotlib.image as mpimg


#Open existing image
im = Image.open('me.jpg')


blurImage = im.filter(ImageFilter.BLUR)
#Save blurImage
blurImage.save('newBlurred.jpg')

# gaussian
BLUR_VALUE = 500
gaussianImage = im.filter(ImageFilter.GaussianBlur(BLUR_VALUE))
#Save blurImage
gaussianImage.save('newGaussian_' +str(BLUR_VALUE)+" TYPE.jpg")


plt.imshow(gaussianImage)
```
