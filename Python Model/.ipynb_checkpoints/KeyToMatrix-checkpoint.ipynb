```python
#Imports
import time
import random
import cv2
import os
import math
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Cursor
from PIL import Image
import hashlib
get_ipython().run_line_magic('matplotlib', 'inline')
```

```python
#Precision 
# exp32 = 4294967296
expValue = 68719476735  #=2^36-1
#Iterations of Cascade
n = 20


originalKey =   0x12d6870025ad0e01d32ec40096b438074cbb1
```

```python
def PL_PWLCM(y0,p2):
    y1=0
    if y0<p2:
        y1 = math.floor(expValue*y0/p2)
    elif p2<y0 and y0 < expValue/2:
        y1 = math.floor(expValue*(y0-p2)/(expValue/2 - p2))
    elif y0 == expValue/2 :
        y1 = 0
    else:
        y1 = PL_PWLCM(expValue-y0, p2)
    return y1

def PL_Logistic(x0):
    x1 = math.floor(4*x0*(expValue-x0)/expValue)
    return x1


def PWLCM_Init(x0,y0,p1,p2):
    x1 = x0
    y1 = y0
    for i in range(n):
#         print('\tIn PWLCM Init for R',i,' and y1 = ', x1, ' z1=', y1)
        x1 = PL_PWLCM(x1,p1)
        y1 = PL_PWLCM(y1,p2)
    return x1,y1

def Logistic_Init(x0):
    x1 = x0
    for i in range(n):
#         print('\tIn logistic Init for R-',i,' and x1 = ', x1)
        x1 = PL_Logistic(x1)
    return x1
```

```python
def getParas(key):
    
    p2 = key & 0x7fffffff
    z0 = (key>>31) & 0x7fffffff
    p1 = (key>>62) & 0x7fffffff
    y0 = (key>>93) & 0x7fffffff
    x0 = (key>>124) & 0xffffffff
    
    print('Paras justr after s[liitinh are x1=',x0,'y=',y0, 'p1=',p1, 'z1=',z0, 'p2=',p2)

    #Split the key in 5 parts
    x0,y0,z0 = InitPRNG(x0,y0,p1,z0,p2)    
    
    ##-> p1,p2 doesnot change
    #p2 is the last 31 bits
    #p1 is the third lat 31 bits
    print('p1 = ', p1, ' and p2 = ', p2)
    return x0,y0,p1,z0,p2


def InitPRNG(x0,y0,p1,z0,p2):
    
    x1 = x0
    y1 = y0
    z1 = z0
    
    for i in range(n):
        y1,z1 = PWLCM_Init(y1,z1,p1,p2)
        x1 = Logistic_Init(x1)
        print('After ',i, 'iteration, x = ', x1, ' y is ', y1, ' and z is ', z1)
        
    return x1,y1,z1


def PL_PWLCM_Logistic(x0,y0,p1,z0,p2):
    
    y1 = PL_PWLCM(y0,p1)
    z1 = PL_PWLCM(z0,p2)
    x1 = PL_Logistic(x0)
    temp1 = x1 ^ y1
    r1 = (temp1 + z1) % expValue
    return x1,y1,z1,r1
```

```python
# get the second column
def takeSecond(elem):
    return elem[1]


def getPRNG(a,b,c,x0,y0,p1,z0,p2):
    
    iLen = math.ceil(a*b*c/4)
    
    sortLength = (a//2)*(b//2)*c 
    
    
    PermutationList = list();
    
    time_start=time.time()
    
    for iNum in range(iLen):
        
        x0,y0,z0,res = PL_PWLCM_Logistic(x0,y0,p1,z0,p2)
        print('iNum = ',iNum,'got are x1=',x0,'y=',y0,  'z1=',z0, 'res=',res)

        
        if iNum < sortLength:
            PermutationList.append([iNum,res])
            

    time_end=time.time()
    
    #Sort the Permutation in increasing iNum
    PermutationList.sort(key=takeSecond)
    
            
                        

    return PermutationList


# After  1 iteration, x =  42010942196  y is  45911045306  and z is  45911045306
# After  0 iteration, x =  42010942196  y is  45911045306  and z is  45911045306

x0,y0,p1,z0,p2 = getParas(originalKey)
print('Paras got are x1=',x0,'y=',y0, 'p1=',p1, 'z1=',z0, 'p2=',p2)
mList= getPRNG(256,256*4,1,x0,y0,p1,z0,p2)
```
