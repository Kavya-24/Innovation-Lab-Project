```python
#Diffusion Testing
#Imports
import time
import random
import cv2
import os
import math
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Cursor
from PIL import Image
import hashlib
get_ipython().run_line_magic('matplotlib', 'inline')




def doRowColPermutationEncryption(img, rowList, colList):
    
    
    w,h = img.shape
    c =  np.zeros((w,h), dtype=np.int)
    c = img.copy()
    istep = h//2
   
    print('\nBefore permuation\n', img)
    #For the permutation of the rows
    for i in range(len(rowList)):
            #All cols of c and its row i is equal to all cols of img and the rowLits[i] row of img 
            c[i,:] = img[rowList[i],:]
 
    print('\nafter row permuation\n', c)
    
    c1 =  np.zeros((w,h), dtype=np.int)
    c1 = c.copy()
    
    for j in range(len(colList)):
            #All rows of c and its col j is equal to all rows of img and the colList[i] row of img 
            c1[:,j] = c[:,colList[j]]
     
  
    
    print('\nafter row col\n', c1)
    return c1


def doRowColPermutationDecryption(img, rowList, colList):
    
    w,h = img.shape
    c =  np.zeros((w,h), dtype=np.int)
    c = img.copy()
    istep = h//2
   
    print('\nBefore dedpermuation\n', img)
    for j in range(len(colList)):
            #All rows of c and its col j is equal to all rows of img and the colList[i] row of img 
            c[:,colList[j]] = img[:,j]
  
    print('\nafter col de permuation\n', c)
    c1 =  np.zeros((w,h), dtype=np.int)
    c1 = img.copy()
    
    #For the permutation of the rows
    for i in range(len(rowList)):
            #All cols of c and its row i is equal to all cols of img and the rowLits[i] row of img 
            c1[rowList[i],:] = c[i,:]
 
  
            
    print('\nafter per\n', c1)
    
    return c1

```
